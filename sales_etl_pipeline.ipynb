```python
import pandas as pd

df = pd.read_csv("/content/raw_sales.csv")
df
```

```python
# Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Convert amount column to numeric
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')

# Remove failed transactions
df_clean = df[df['status'] == 'Completed']

# Fill missing amount with 0
df_clean['amount'] = df_clean['amount'].fillna(0)

df_clean

```

```python
payment_summary = (
    df_clean
    .groupby('payment_mode')['amount']
    .sum()
    .reset_index()
)

payment_summary
```

```python
df_clean.to_csv("processed_sales.csv", index=False)
payment_summary.to_csv("payment_summary.csv", index=False)

print("ETL pipeline completed successfully")
```
